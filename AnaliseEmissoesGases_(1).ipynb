# Ajustando a base de dados
Estamos interessados somente na emissão de gases gerados pelos estados que estejam no investário nacional, portanto vamos identificar tipos de dados únicos da coluna Emissão / Remoção / Bunker.

* Remoção são dados referentes a retirada dos gases estufa, que é o contrário da emissão.
* NCI são daods não contemplados no inventário nacional
* Bunkers são emissões de transporte marítimo e aéreo internacional, portanto não são emissões diretas dos estados.

Vamos checar primeiro os dados de remoção, para identificar se todos eles são de retirada de gases efeito estufa.


```python
emissoes_gases['Emissão / Remoção / Bunker'].unique()
```

Método ISIN para selecionar dois ou mais valores dentro de uma coluna.


```python
(emissoes_gases['Emissão / Remoção / Bunker'] == 'Remoção NCI') | (emissoes_gases['Emissão / Remoção / Bunker'] == 'Remoção')
```

```python
emissoes_gases[emissoes_gases['Emissão / Remoção / Bunker'].isin(['Remoção NCI', 'Remoção'])]
```

Método LOC para selecionar linhas e colunas


```python
emissoes_gases.loc[emissoes_gases['Emissão / Remoção / Bunker'].isin(['Remoção NCI', 'Remoção']), 1970 : 2021]
```

```python
emissoes_gases.loc[emissoes_gases['Emissão / Remoção / Bunker'].isin(['Remoção NCI', 'Remoção']), 1970 : 2021].max()
```

Vamos tambem identificar se algum valor do tipo Bunker corresponde a alguma emissão feita por algum estado.

```python
emissoes_gases.loc[emissoes_gases['Emissão / Remoção / Bunker'] == 'Bunker', 'Estado'].unique()
```

```python
emissoes_gases = emissoes_gases[emissoes_gases['Emissão / Remoção / Bunker']== 'Emissão']
emissoes_gases
```

Como não restou nenhum valor útil para análise da coluna Emissão / Remoção / Bunker, vamos remove-la por completo da base de dados

```python
emissoes_gases = emissoes_gases.drop(columns = 'Emissão / Remoção / Bunker')
emissoes_gases
```

# Modificando o formato do DataFrame


Para realizaer as agregações nos dados de emissão, é necessário que a base de dados esteja em um formato adequado, com os valores de emissão todos em uma mesma coluna. Para isso, devemos modificar o formato da base de dados de forma que as colunas referentes a cada ano se tornem uma só coluna.

```python
emissoes_gases.loc[:, 'Nível 1 - Setor' : 'Produto'].columns
```

```python
colunas_info = list(emissoes_gases.loc[:, 'Nível 1 - Setor' : 'Produto'].columns)
colunas_info
```

```python
emissoes_gases.loc[:, 1970:2021].columns
```

```python
colunas_emissao = list(emissoes_gases.loc[:, 1970:2021].columns)
colunas_emissao
```

Método melt permite pegar várias colunas e transformar apenas em uma.
Para realizar a transformação do DataFrame, utilizamos o método melt(). Esse método tem o intuito de transformar um DataFrame de um formato amplo (wide) para o formato longo (long).


```python
emissoes_gases.melt(id_vars = colunas_info, value_vars = colunas_emissao, var_name = 'Ano', value_name = 'Emissão')
```

```python
emissoes_por_ano = emissoes_gases.melt(id_vars = colunas_info, value_vars = colunas_emissao, var_name = 'Ano', value_name = 'Emissão')
```
